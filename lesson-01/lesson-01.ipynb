```python
n = 10 ** 7
k = 10 ** 6
lst = list(range(0, n))
```

```python
def naive_search(lst, k):
    # O(n)
    op_cnt = 0
    for num in lst:
        op_cnt += 1
        if num == k:
            print(f'Count of operations is {op_cnt}')
            return num
    print(f'Count of operations is {op_cnt}')
    return None

# [2]
# [1,2,3,4,5,6,7,8,9]
# k = 7
def binary_search(lst, k):
    op_cnt = 0
    low = 0
    high = len(lst) - 1
    while low <= high:
        op_cnt += 1
        mid = (low + high) // 2
        guess = lst[mid]
        if guess == k:
            print(f'Count of operations is {op_cnt}')
            return guess
        if k > guess:
            low = mid + 1
        else: 
            high = mid - 1
    print(f'Count of operations is {op_cnt}')
    return None
```

```python
#%%timeit
naive_search(lst, k)
```

```python
#%%timeit
binary_search(lst, k)
```
